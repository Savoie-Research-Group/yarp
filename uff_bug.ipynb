```python
import yarp as yp
from yarp.reaction.enum import enumerate_products
from yarp.reaction.generate_rxns import quick_geom_opt

reactant = yp.yarpecule("reactant.xyz")
products = list(enumerate_products(reactant, 2, 2, mode="concerted"))

target_hash = "1034502.7961211"

target_product = next(p for p in products if str(p.hash) == str(target_hash))

opt_product = quick_geom_opt(target_product)

reinit_product = yp.yarpecule(
    (
        opt_product.adj_mat.copy(),
        opt_product.geo.copy(),
        list(opt_product.elements),
        int(opt_product.q),
        {i: dict(opt_product._atom_info[i]) for i in opt_product._atom_info},
    ),
    canon=False,
)
print(reinit_product._atom_info)
print("Hash:", reinit_product.hash)
print("Geometry:")
print(reinit_product.geo)

reinit_product.draw_bmats(show_inline=True)
```
