```python
def pretty_print(elements, adj_mat):
    # Set column width for alignment
    col_width = 5
    # Print header row with element symbols
    header = " " * col_width + "".join(f"{el:>{col_width}}" for el in elements)
    print(header)
    # Print each row: element symbol + adjacency values
    for el, row in zip(elements, adj_mat):
        row_str = "".join(f"{val:>{col_width}}" for val in row)
        print(f"{el:>{col_width}}{row_str}")

def print_mats(smi, mode):
    mol = yp.yarpecule(smi, mode=mode)

    print(f"-***- {smi} with {mode} -***-")
    print("")

    print(f"Adjacency matrix:")
    pretty_print(mol.elements, mol.adj_mat)
    print("")

    for i, be in enumerate(mol.lewis.bond_mats):
        print(f"BE matrix #{i}:")
        pretty_print(mol.elements, be)
        print("")
```

```python
ethene_smi = "C=C"
formaldehyde_smi = "C=O"
aro_smi = "n1ccccc1"
```

## Here's what RDKit mode gives us

```python
print_mats(ethene_smi, "rdkit")
print_mats(formaldehyde_smi, "rdkit")
print_mats(aro_smi, "rdkit")
```

## And here's what YARP mode gives us

```python
print_mats(ethene_smi, "yarp")
print_mats(formaldehyde_smi, "yarp")
print_mats(aro_smi, "yarp")
```

## Onwards to product enumeration!

```python
from yarp.reaction.enum import break_bonds, form_n_bonds
import numpy as np
```

### Diels-Alder bimolecular test case

```python
xyz_path = "/Users/erm/git_stuff/yarp-again/test/molecules/diels_alder.xyz"
da_mol = yp.yarpecule(xyz_path)
pretty_print(da_mol.elements, da_mol.adj_mat)
```

```python
break_da = list(break_bonds(da_mol, n=2))
print(f"We have {len(break_da)} Diels-Alder intermediates after breaking 2 bonds.")
```

```python
prod_da = form_n_bonds(break_da, n=2)
print(f"We have {len(prod_da)} Diels-Alder potential products after forming 2 bonds.")

prod_da = [_ for _ in prod_da if _.bond_mat_scores[0]<=0.0 and sum(np.abs(_.fc) <= 2.0)]

print(f"We have {len(prod_da)} Diels-Alder potential products after filtering.")
```
