```python
#Load in 1-methyl-3-buteneium as a yarpecule and draw its bond matrix
y_i = yp.yarpecule(ROOT + FILE)

y_i.draw_bmats(show_inline=True)
```

Print all the internal attributes of the yarpecule

```python
print(f"File Name: {FILE}")
print(f"Elements: {y_i.elements}")
print(f"Formal Charge: {y_i.fc}")
print(f"Scores: {y_i.bond_mat_scores}")
print(f"Rings:\n{y_i.rings}")
print(f"E_Accepters: {y_i.n_e_accept}")
print(f"E_Donors: {y_i.n_e_donate}")
print(f"atom_neighbors: {y_i.atom_neighbors}")
print(f"Bond Mat:\n{y_i.bond_mats}")
print(f"Charge: {y_i.q}")
```

```python
print("Formal Charges with Indexes:")
for idx, fc in enumerate(y_i.fc):
    print(f"assert yp_mol.fc[{idx}] == {fc}")
```

```python
print("Bond Matrices Scores with Indexes:")
for idx, score in enumerate(y_i.bond_mat_scores):
    print(f"assert yp_mol.bond_mat_scores[{idx}] == {score}")
```

```python
print("Rings with Indexes:")
for idx, ring in enumerate(y_i.rings):
    for r_idx, r in enumerate(ring):
        print(f"assert yp_mol.rings[{idx}][{r_idx}] == {r}")
    
```

```python
print("E Acceptors with Indexes:")
for idx, e_acceptor in enumerate(y_i.n_e_accept):
    print(f"assert yp_mol.n_e_accept[{idx}] == {e_acceptor}")
```

```python
print("E Donors with Indexes:")
for idx, e_donor in enumerate(y_i.n_e_donate):
    print(f"assert yp_mol.n_e_donate[{idx}] == {e_donor}")
```

```python
print("Atom Neighbors with Indexes:")
for idx, neighbors in enumerate(y_i.atom_neighbors):
    print(f"assert yp_mol.atom_neighbors[{idx}] == {neighbors}")
```

```python
print("Bond Matrix Elements with Indexes:")
print(f"assert len(yp_mol.bond_mats) == {len(y_i.bond_mats)}")
for bmat_idx, bond_mat in enumerate(y_i.bond_mats):
    print(f"yarpecule.bond_mats[{bmat_idx}]")
    for row_idx, row in enumerate(bond_mat):
        print(f"    yarpecule.bond_mats[{bmat_idx}][{row_idx}]")
        for column_idx, column in enumerate(row):
            print(f"        assert yp_mol.bond_mats[{bmat_idx}][{row_idx}][{column_idx}] == {column}")
```
