```python
print("Yarpecule Charge:", y_i.q)
```

```python
print("Elements with Indexes:")
for idx, elem in enumerate(y_i.elements):
    print(f"yarpecule.elements[{idx}] = {elem}")
```

```python
print("Formal Charges with Indexes:")
for idx, fc in enumerate(y_i.fc):
    print(f"yarpecule.fc[{idx}] = {fc}")
```

```python
print("Bond Matrices Scores with Indexes:")
for idx, score in enumerate(y_i.bond_mat_scores):
    print(f"yarpecule.bond_mat_scores[{idx}] = {score}")
```

```python
print("Rings with Indexes:")
for idx, ring in enumerate(y_i.rings):
    print(f"yarpecule.rings[{idx}] = {ring}")
```

```python
print("E Acceptors with Indexes:")
for idx, e_acceptor in enumerate(y_i.n_e_accept):
    print(f"yarpecule.n_e_accept[{idx}] = {e_acceptor}")
```

```python
print("E Donors with Indexes:")
for idx, e_donor in enumerate(y_i.n_e_donate):
    print(f"yarpecule.n_e_donate[{idx}] = {e_donor}")
```

```python
print("Atom Neighbors with Indexes:")
for idx, neighbors in enumerate(y_i.atom_neighbors):
    print(f"yarpecule.atom_neighbors[{idx}] = {neighbors}")
```

```python
print("Bond Matrix Elements with Indexes:")
for bmat_idx, bond_mat in enumerate(y_i.bond_mats):
    print(f"yarpecule.bond_mats[{bmat_idx}]")
    for row_idx, row in enumerate(bond_mat):
        print(f"    yarpecule.bond_mats[{bmat_idx}][{row_idx}]")
        for column_idx, column in enumerate(row):
            print(f"        yarpecule.bond_mats[{bmat_idx}][{row_idx}][{column_idx}] = {column}")
```

```python
print("Adjacency Matrix with indexes, and their indexes:")
for row_idx, row in enumerate(y_i.adj_mat):
    print(f"yarpecule.adj_mat[{row_idx}]")
    for column_idx, column in enumerate(row):
        print(f"  yarpecule.adj_mat[{row_idx}][{column_idx}] = {column}")
```
